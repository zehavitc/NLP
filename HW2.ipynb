```
import random
from collections import Counter, defaultdict

class simple_tagger:
  
  
  def train(self, data):
    d = defaultdict(Counter)
    self.possible_tags = set()
    for sentence in data:
      for t in sentence:
        d[t[0]][t[1]] += 1 
        self.possible_tags.add(t[1])
    #find max tag for each word
    self.map_tags = dict()
    for word in d.keys():
      self.map_tags[word] = random.choice(d[word].most_common(1))[0]
    self.possible_tags = list(self.possible_tags)
    print(self.map_tags)
    print(self.possible_tags)
      
  
  def evaluate(self, data):
    total_words = 0
    success_words = 0
    total_sent = 0
    succss_sent = 0
    for sentence in data:
      #print(sentence)
      total_sent += 1
      is_correct = True
      for word in sentence:
        total_words += 1
        train_tag = self._get_tag(word[0])
        if train_tag==word[1]:
          #print(f'success word {word}')
          success_words += 1
        else:
          #print(f'worng word {word}')
          is_correct = False
      if is_correct:
        succss_sent += 1
    
    word_accuracy = success_words / total_words
    sent_accuracy = succss_sent / total_sent
    return (word_accuracy, sent_accuracy)
  
  def _get_tag(self, word):
    if word in self.map_tags:
      return self.map_tags[word]
    secure_random = random.SystemRandom()
    return random.choice(self.possible_tags)
```

```
import numpy as np
import operator


def viterbi (word_list, A, B, Pi):

    # initialization
    T = len(word_list)
    N = A.shape[0] # number of tags

    delta_table = np.zeros((N, T)) # initialise delta table
    psi = np.zeros((N, T))  # initialise the best path table

    delta_table[:,0] = B[:, word_list[0]] * Pi

    for t in range(1, T):
        for s in range (0, N):
            trans_p = delta_table[:, t-1] * A[:, s]
            psi[s][t], delta_table[s][ t] = max(enumerate(trans_p), key=operator.itemgetter(1))
            delta_table[s][t] = delta_table[s][t] * B[s][word_list[t]]

    # Back tracking
    seq = np.zeros(T)
    seq[T-1] = delta_table[:, T-1].argmax()
    for t in range(T-1, 0, -1):
        seq[t-1] = psi[int(seq[t])][t]

    return seq



# A = np.array([[0.3, 0.7], [0.2, 0.8]])
# B = np.array([[0.1, 0.1, 0.3, 0.5], [0.3, 0.3, 0.2, 0.2]])
# Pi = np.array([0.4, 0.6])
# print(viterbi([3, 3, 3, 3], A, B, Pi))
```

```
class hmm_tagger:
    
  def train(self,data):
    word_idx = 0
    self.map_words = dict()
    self.map_idx_to_word = dict()
    self.map_tags = dict()
    self.map_idx_to_tag = dict()
    
    count_tag_transitions = defaultdict(Counter)
    count_word_tags = defaultdict(Counter)
    count_tags = Counter()
    tags_count = 0
    
    tag_idx = 0
    for sent in data:
      for t in sent:
        if t[1] not in self.map_tags:
          self.map_tags[t[1]] = tag_idx
          self.map_idx_to_tag[tag_idx] = t[1]
          tag_idx += 1
    
    for sent in data:
      prev_t = None
      for t in sent:
        if (t[0] not in self.map_words):
          self.map_words[t[0]] = word_idx
          self.map_idx_to_word[word_idx] = t[0]
          word_idx += 1
        count_tags[t[1]] += 1
        count_word_tags[t[1]][self.map_words[t[0]]] += 1
        if (prev_t != None):
          count_tag_transitions[prev_t][t[1]] += 1         
        prev_t = t[1]
        tags_count += 1
    self.A = []
    num_tags = len(self.map_tags)
    for tag_idx in range(num_tags):
      tag_list = [0] * num_tags
      tag = self.map_idx_to_tag[tag_idx]
      for next_tag_idx in range(num_tags):        
        tag_list[next_tag_idx] = count_tag_transitions[tag][self.map_idx_to_tag[next_tag_idx]] / count_tags[tag]
      self.A.append(tag_list)
    self.pi = []
    for tag_idx in range(num_tags):
      self.pi.append(count_tags[self.map_idx_to_tag[tag_idx]] / tags_count)
    self.B =  []
    num_words = len(self.map_words.keys())
    for tag in count_word_tags.keys():
      tag_list = [0] * num_words
      for word_idx in range(num_words):                                  
        tag_list[word_idx] = (count_word_tags[tag][word_idx] / count_tags[tag])
      #print(tag_list)
      self.B.append(tag_list)
     
    self.A  = np.array(self.A)
    self.B = np.array(self.B)
    self.pi = np.array(self.pi)
#     print(self.A)
#     print("------------------------")
#     print(self.B)
#     print("------------------------")
#     print(self.pi)
#     print(sum(self.B[0]))

  
  def evaluate(self, data):
    results = []
    possible_tags = list(self.map_idx_to_tag.keys())
    for sent in data:
      obs = []
      sent_result = []
      for t in sent:
        if t[0] not in self.map_words.keys():
          if (len(obs) != 0):
            sent_result.extend(viterbi(obs, self.A, self.B, self.pi))
          secure_random = random.SystemRandom()
          sent_result.append(secure_random.choice(possible_tags))
          obs = []
        else:
          obs.append(self.map_words[t[0]])    
      if (len(obs) != 0):
        sent_result.extend(viterbi(obs, self.A, self.B, self.pi))
      results.append(sent_result)
      
    # calc the accuracy 
    total_words = 0
    success_words = 0
    total_sent = 0
    succss_sent = 0
    for sent_idx in range(len(data)):
      is_correct = True
      word_idx_in_sent = 0
      for t in data[sent_idx]:
        real_tag = self.map_tags[t[1]]
        tag = results[sent_idx][word_idx_in_sent]
        if (tag == real_tag):
          success_words  += 1
        else:
          is_correct = False
        total_words += 1
        word_idx_in_sent += 1
      total_sent += 1
      if (is_correct):
        succss_sent += 1
    
    word_accuracy = success_words / total_words
    sent_accuracy = succss_sent / total_sent
    return (word_accuracy, sent_accuracy) 
        
      
    
```

```
tagger = hmm_tagger()
tagger.train(train_data)
print(tagger.evaluate(test_data))
```
